```python
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data_for_predictions.csv')
df
```

```python
df = df.drop(columns=['Unnamed: 0', 'id'])
```

```python
df.columns
```

```python
x=df.drop(['churn'],axis=1)
y=df.churn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
RF=RandomForestClassifier(n_estimators=100,random_state=10)
```

```python
RF.fit(x_train,y_train)
```

```python
y_pred=RF.predict(x_test)
y_pred
```

```python
predicted_df = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred})
predicted_df
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)
```

```python
from sklearn.metrics import roc_auc_score

# Assuming y_pred contains predicted probabilities for binary classification
# and x_test contains the corresponding features
auc_score = roc_auc_score(y_test, y_pred)
print("AUC Score:", auc_score)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='viridis', xticklabels=df.churn.unique(), yticklabels=df.churn.unique())
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split

# Define the hyperparameter grid to search
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Use GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(estimator=RF, param_grid=param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

# Print the best hyperparameters
print("Best Hyperparameters:", grid_search.best_params_)

# Get the best model
best_model = grid_search.best_estimator_

# Make predictions on the testing data using the best model
y_pred = best_model.predict(x_test)

conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

# Evaluate the performance of the best model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of the Best Model: {accuracy:.2f}')

from sklearn.metrics import roc_auc_score

# Assuming y_pred contains predicted probabilities for binary classification
# and x_test contains the corresponding features
auc_score = roc_auc_score(y_test, y_pred)
print("AUC Score:", auc_score)
```

```python
# Get feature importances
importances = RF.feature_importances_

# Get the indices of features sorted by importance
indices = sorted(range(len(importances)), key=lambda i: importances[i], reverse=True)

# Print feature ranking with feature names
print("Feature ranking:")
for i in range(len(importances)):
    print(f"Feature {indices[i] + 1}: {df.columns[indices[i]]} - Importance: {importances[indices[i]]}")

# Plot the feature importances with adjusted figure size
plt.figure(figsize=(30, 6))  # Adjust the width and height of the figure as needed
plt.title("Feature importances")
plt.bar(range(x.shape[1]), importances[indices], align="center")
plt.xticks(range(x.shape[1]), [df.columns[i] for i in indices], rotation=45)
plt.xlabel("Feature")
plt.ylabel("Feature importance")
plt.show()
```
